<h1> <center> Introdução ao Equilíbrio Líquido/Vapor

O equilíbrio líquido/vapor refere-se a um estado estático onde não há mudanças macroscópicas no sistema ao longo do tempo. Isso significa que as propriedades como temperatura, pressão e composição das fases líquida e vapor não variam. Embora haja movimento molecular contínuo e trocas entre as fases, as taxas médias de transferência são iguais em ambas as direções, resultando em um sistema balanceado sem transferência líquida de matéria.

A composição de misturas ou soluções pode ser expressa de três formas principais:



<h2> <u>Regra das Fases</u>


A Regra das Fases descreve a relação entre as variáveis de um sistema termodinâmico em equilíbrio sem reações químicas. A fórmula apresentada é:

#### $$\boxed { F = 2 + (N - 1)\pi - (\pi - 1)N }$$

onde:
- $ F $ é o número de graus de liberdade do sistema (número de variáveis que podem ser alteradas independentemente sem afetar outras variáveis),
- $ N $ é o número de componentes no sistema,
- $ \pi $ é o número de fases em equilíbrio.

Simplificando a expressão, obtemos a regra das fases como:

#### $$\boxed { F = 2 - \pi + N }$$

Isto significa que o número de condições que podemos especificar livremente em um sistema é igual a dois mais o número de componentes menos o número de fases.



<h2> <u>Teorema de Duhem</u>


Para qualquer sistema fechado formado inicialmente por espécies químicas com quantidades de massa especificadas, o estado de equilíbrio é completamente determinado quando são fixadas quaisquer duas variáveis independentes.

As variáveis independentes podem ser intensivas (como temperatura e pressão) ou extensivas (como volume e quantidade de substância). Quando o grau de liberdade $ F $ é 1, pelo menos uma das variáveis deve ser extensiva, e quando $ F $ é 0, todas as variáveis devem ser extensivas.


<h2> <u>Comportamento Qualitativo do ELV:</u> Sistemas Binários


No equilíbrio líquido/vapor (ELV) de sistemas com duas espécies químicas, podemos analisar o comportamento das fases usando um diagrama tridimensional P-T-composição. Esse espaço tridimensional é definido pelas variáveis de pressão (P), temperatura (T) e composição (fração molar X ou Y). O grau de liberdade (F) é dado pela regra das fases para sistemas binários, $ F = 4 - \pi $, onde $ \pi $ é o número de fases presentes.

### Superfícies no Diagrama P-T-Composição:



- **Superfície Inferior (P-T-Y1)**: Representa o estado de vapor saturado.
- **Superfície Superior (P-T-X1)**: Representa o estado de líquido saturado.

Essas superfícies se interceptam ao longo das curvas que representam a pressão de vapor versus temperatura para as espécies puras, formando uma superfície contínua na parte superior do diagrama.

#### Pontos e Regiões Críticas:

- **Pontos Críticos (C1 e C2)**: São os pontos nas superfícies onde as fases líquida e vapor se tornam idênticas. Eles se encontram no extremo arredondado da superfície contínua entre C1 e C2.
- **Região do Líquido Sub-resfriado**: Localizada acima da superfície superior.
- **Região do Vapor Superaquecido**: Localizada abaixo da superfície inferior.

#### Dinâmica do Equilíbrio:

- **Ponto de Bolha (L)**: O ponto onde a primeira bolha de vapor aparece a partir de um líquido em equilíbrio, sob redução de pressão.
- **Ponto de Orvalho (W)**: O ponto onde as últimas gotas de líquido desaparecem, representando um estado de vapor saturado.

O interior entre as duas superfícies representa a coexistência das fases líquida e vapor. As linhas de amarração no diagrama unem pontos representando fases em equilíbrio.

### Representação Bidimensional:

Devido à complexidade dos sistemas, muitas vezes, o ELV binário é representado por gráficos bidimensionais que cortam o diagrama tridimensional. Estes gráficos, como o mostrado na Figura 10.2(a), apresentam as relações P-x1-y1 para diferentes temperaturas, ilustrando o comportamento do sistema a uma temperatura constante. Com isso tambem tem-se o contrário na Figura 10.2(b) para as diferentes pressões constantes.





<h3> <u> Comportamento de Fases em Sistemas Binários</u></h3>



- **Diagramas T-x-y e P-x-y**: Representam a relação entre temperatura (T), pressão (P), e as composições da fase líquida (x) e da fase vapor (y) para misturas binárias. 

- **Ponto Crítico**: O ponto crítico em um diagrama P-T ocorre onde a curva de envelope é tangente à curva crítica. Esse ponto representa a temperatura e pressão mais altas em que as fases líquida e vapor podem coexistir.
- **Curvas de Envelope**: As curvas em um diagrama P-T para diferentes composições mostram como a pressão de vapor varia com a temperatura para composições específicas de uma mistura binária.

<h3> <u> Condensação Retrógrada</u></h3>



- **Comportamento da Mistura Perto do Ponto Crítico:**
    - **Vaporização (Líquido para Vapor):** À esquerda do ponto crítico, quando a pressão é reduzida, a mistura segue um caminho do ponto de bolha ao ponto de orvalho, resultando na vaporização esperada do líquido.
    - **Condensação Retrógrada (Vapor para Líquido):** Se a mistura começar como vapor saturado (ponto F), a redução da pressão leva primeiro à condensação até o ponto G, onde a composição do líquido atinge um máximo. Após isso, a mistura se move em direção a uma vaporização retrógrada até o ponto de orvalho (ponto H).

- **Azeótropos**: Pontos no diagrama onde a composição da fase líquida é igual à da fase vapor. Azeótropos podem ser de mínimo ou máximo, dependendo se representam um ponto de ebulição máxima ou mínima.

<h2> <u> Lei de Raoult</u>


A Lei de Raoult é baseada em duas hipóteses fundamentais:

1. **Fase Vapor como Gás Ideal**
2. **Fase Líquida como Solução Ideal**

A Lei de Raoult relaciona a pressão parcial de um componente no vapor ($ y_i P $) com a sua fração molar na fase líquida ($ x_i $) e sua pressão de vapor saturado ($ P_i^{sat} $), dada pela equação:

#### $$\boxed { y_i P = x_i P_i^{sat} \quad (i = 1, 2, ..., N)}$$


<h3> <u> Cálculos de Ponto de Bolha e Orvalho</u></h3>

Os cálculos de ponto de bolha e orvalho são essenciais para engenheiros químicos, pois permitem determinar as condições de temperatura e pressão sob as quais uma mistura começa a ferver ou condensar. Existem quatro classes de cálculos baseados na Lei de Raoult:

1. **BOL P**: Cálculo de $ y_i $ e $ P $ dados $ x_i $ e $ T $.
2. **ORV P**: Cálculo de $ x_i $ e $ P $ dados $ y_i $ e $ T $.
3. **BOL T**: Cálculo de $ y_i $ e $ T $ dados $ x_i $ e $ P $.
4. **ORV T**: Cálculo de $ x_i $ e $ T $ dados $ y_i $ e $ P $.

Para sistemas binários, as equações da Lei de Raoult podem ser reorganizadas para expressar a pressão total $ P $ como:

#### $\boxed { P = \sum x_i P_i^{sat} \quad \text{(para ponto de bolha)}}$

#### $\boxed { P = \left( \sum \frac{y_i}{P_i^{sat}} \right)^{-1} \quad \text{(para ponto de orvalho)}}$

onde $ x_i $ é a fração molar na fase líquida e $ y_i $ é a fração molar na fase vapor.


<h2> <u>Lei de Henry</u>


A Lei de Henry é útil para calcular a solubilidade de gases em líquidos em condições onde o gás pode ser considerado ideal. Esta lei afirma que a pressão parcial do gás $( y_i P $) é diretamente proporcional à sua fração molar na fase líquida $( x_i $), expressa pela equação:

### $$\boxed { y_i P = x_i H_i }$$

onde $ H_i $ é a constante de Henry para a espécie $ i $ e pode ser encontrada experimentalmente. Os valores de $ H_i $ são fornecidos para várias espécies a uma temperatura específica, geralmente 25°C.

### Aplicação da Lei de Henry



Como exemplo, o texto descreve um cenário em que se deseja calcular a fração molar de ar dissolvido na água a 25°C e pressão atmosférica. A Lei de Raoult não seria adequada para o ar devido à sua baixa temperatura crítica. Neste caso, a Lei de Henry pode ser aplicada com a seguinte adaptação:

### $$\boxed { x_1 = \frac{y_1 P}{H_1}} $$

Tomando $ y_1 $ como $ 1 - y_2 $ (assumindo que a água é a outra espécie e que sua fração molar é $ y_2 = 0,0312$), e aplicando os valores conhecidos de pressão atmosférica (101,33 kPa) e a constante de Henry para o ar (72.950 $ \frac{l}{\text{bar}} $), é possível calcular a fração molar de ar na água. O cálculo resulta em:

#### $$\boxed { x_1 = \frac{(0,9688)(101,33)}{72.950} = 1,35 \times 10^{-5} }$$

Esse resultado valida o uso da Lei de Raoult para o componente da água, supondo que a água é a espécie pura e que a Lei de Raoult é aplicável para ela sob as condições dadas.

<h2> <u>Lei de Raoult Modificada</u>



A Lei de Raoult Modificada é uma adaptação da Lei de Raoult que incorpora coeficientes de atividade ($ \gamma_i $) para contabilizar os desvios da idealidade em soluções reais:

#### $$\boxed { y_i P = \gamma_i x_i P_i^{sat} }$$

Aqui, $ y_i $ é a fração molar na fase vapor, $ \gamma_i $ é o coeficiente de atividade, $ x_i $ é a fração molar na fase líquida, e $ P_i^{sat} $ é a pressão de vapor saturada da espécie $ i $ pura.

<h3> <u> Cálculos de Ponto de Bolha e Orvalho</u></h3>

#### $\boxed { P = \sum \gamma_i x_i P_i^{sat} \quad \text{(para ponto de bolha)}}$

#### $\boxed { P = \left( \sum \frac{y_i}{\gamma_i P_i^{sat}} \right)^{-1} \quad \text{(para ponto de orvalho)}}$


<h2> <u>Correlações para o Valor K</u>


O valor K é uma razão de equilíbrio que indica a tendência de uma espécie química em *dividir-se entre as fases líquida e vapor*. É definido como:

### $$\boxed { K_i = \frac{y_i}{x_i} }$$

- **Lei de Raoult:**

  - #### $\boxed { K_i = \frac{P_i^{sat}}{P} }$

- **Lei de Raoult Modificada:**

  - #### $\boxed { K_i = \frac{\gamma_i P_i^{sat}}{P} }$

Onde $ P $ é a pressão total do sistema.

#### $\boxed { \sum K_i x_i = 1 }\quad \text{(para ponto de bolha)}$

#### $\boxed { \sum \frac{y_i}{K_i} = 1 }\quad \text{(para ponto de orvalho)}$


<h2> <u>Cálculos de Flash </u>


No processo de flash, um líquido é parcialmente evaporado ao reduzir a pressão abaixo da sua pressão no ponto de bolha, resultando em um sistema bifásico de vapor e líquido em equilíbrio. Este cálculo é conhecido como flash-P,T, que se concentra nas grandezas e composições das fases vapor e líquida em equilíbrio a uma temperatura (T), pressão (P), e composição global conhecida. O problema é resolvido aplicando o teorema de Duhem, que permite definir o sistema com duas variáveis independentes (T e P) para uma composição fixa.

### Dedução

Consideremos um sistema com um mol de espécies químicas que não reagem, com uma composição global dada pelas frações molares $ z_i $. Se $ L $ for a quantidade de líquido com frações molares $ x_i $ e $ V $ a quantidade de vapor com frações molares $ y_i $,Portanto temos que **$L + V =1$** e as equações de balanço de massa são expressas como:

#### $\boxed {  z_i = x_iL + y_iV} \quad (i = 1, 2, ..., N)$

Para eliminar $ L $, combinamos as equações para obter:

#### $\boxed {  z_i = x_i(1 - V) + y_iV} \quad (i = 1, 2, ..., N) $

Substituindo $ x_i $ por $ y_i/K_i $ (onde $ K_i $ é o coeficiente de distribuição), e resolvendo para $ y_i $, temos:

#### $\boxed {  y_i = \frac{z_iK_i}{1 + V(K_i - 1)}} \quad (i = 1, 2, ..., N) $

Sendo que a soma das frações molares do vapor é igual a 1, a equação é somada para todas as espécies para encontrar $ V $:

#### $\boxed {  \sum_{i=1}^{N} \frac{z_iK_i}{1 + V(K_i - 1)} = 1 }$

O primeiro passo na resolução de um cálculo de flash-P,T é determinar o valor de \( V \) que satisfaz essa última equação. Note-se que $ V = 1 $ é uma solução trivial (onde todo o sistema é vapor).
