<h2 align="center" style="color:gray">Concentrações</h2>

---

#### 1. Concentração ou densidade mássica total (ρ):
massa total do sistema de N componentes por volume da mistura.

$$\rho i = \frac{mi}{V}$$

#### 2. Concentração Molar (Ci): 
número de mols de soluto por unidade de volume de uma mistura com N componentes.

$$Ci = \frac{ni}{V}$$



#### 3. Fração mássica (w):
Relação entre a massa da espécie i e a massa total da mistura

$$\omega i = \frac{mi}{mt}$$
    
#### 4.  Fração molar (x): 
Relação entre o número de mols da espécie i e o número de mols total.
    
 $$ xi = \frac{ni}{n}$$

#### Algumas relações importantes:

$$
\rho i = Ci . Mi \\
\rho i = \omega i . \rho\\
Ci = xi . C\\
M = \sum_{}^{} xi . Mi
$$

<h2 align="center" style="color:gray">Velocidades</h2>

---

#### 5. Velocidade da espécie química ``i`` em relação ao eixo de coordenadas fixo:



#### 6. Velocidade mássica média local da mistura **($v$)**:

$$
    \rho * v = \sum_{}^{} \rho i * vi\\
    v = \sum_{}^{} \frac{\rho i}{\rho} * vi\\
\rightarrow \color{Blue}{v = \sum_{}^{} \omega i * vi}\\
$$



#### 7. Velocidade molar média local da mistura **($v^{*}$)**:

$$
    C * v^{*} = \sum_{}^{} Ci * vi\\ 
    v^{*} = \sum_{}^{} \frac{Ci}{C} * vi\\ 
\rightarrow  \color{Blue}{v^{*} = \sum_{}^{} xi * vi}\\
$$

#### 8. Velocidade mássica de difusão de ``i``:

$$ i = (vi - v)$$

#### 9. Velocidade molar de difusão de ``i``:

$$ i = (vi - v^{*})$$

### Lei de Fick

---

A lei de Fick é uma equação matemática que descreve como a concentração de uma substância se espalha através de um meio.



A equação pode ser escrita de duas maneiras diferentes, dependendo da abordagem usada. A primeira forma é conhecida como a equação da difusão unidimensional, e a segunda forma é conhecida como a equação da difusão tridimensional.

Equação unidimencional no sentido de z

$$
J_{az}^{*} = -D_{ab} \frac{\mathrm{d} Ca}{\mathrm{d} z} = -D_{ab} * C *\frac{\mathrm{d} xa}{\mathrm{d} z}
$$

Tridimensional:
$$
J_{az}^{*} = - D_{ab} . C . \bigtriangledown_{xa}
$$

Lembrando que:

$$
J_{a}^{*} = Ca . (va-v^{*})\\
J_{a}^{*} = Ca.va-Ca.v^{*}\\
J_{a}^{*} = Na-Ca.(xa.va+xb.vb)\\
J_{a}^{*} = Na-Ca.(\frac{Ca}{C}.va+\frac{Cb}{C}.vb)\\
J_{a}^{*} = Na-\frac{Ca}{C}.(Ca.va+Cb.vb) \\
J_{a}^{*} = Na-xa.(Na+Nb)\\
Na = xa.(Na+Nb) + J_{a}^{*} \\
Na = \underbrace{\left(  xa.(Na+Nb) \right)}_{Termo Convectivo}
\underbrace{\left( - D_{ab} . C . \bigtriangledown_{xa} \right)}_{Termo Difusivo}
$$

Analogamente:
$$
j_{a}^{*} = - \rho D_{ab}. \bigtriangledown_{\omega a}\\
na = \omega . (na+nb) - \rho D_{ab}. \bigtriangledown_{\omega a}
$$

<h2 align="center" style="color:gray">Difusividade em liquidos</h2>

A **difusividade em líquidos** é a medida da facilidade com que uma molécula ou um íon se move através de um meio líquido. A difusividade é influenciada por varios fatores, como a natureza das moléculas ou íons, a temperatura e a pressão. Em geral, as ``moléculas menores têm uma difusividade maior do que as moléculas maiores``. A difusividade também aumenta com a temperatura.

#### A difusividade em liquidos está entre:
<h2>$$10^{-6}< D_{ab} < 10^{-5}$$
