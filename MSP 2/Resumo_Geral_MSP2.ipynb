## # Linearização de modelos

### Definição

Em matemática e suas aplicações, linearização refere-se a encontrar uma aproximação linear de uma função em um dado ponto.

No estudo de sistemas dinâmicos, a linearização é um método para avaliar a estabilidade local de um ponto de equilibrio de um sistema de equações diferenciais não lineares.

### Importância da linearização

Na prática, a maioria dos sistemas reais geram modelos não lineares. Trabalhar com modelos não lineares é uma tarefa complicada pois, a maioria das tecnicas anaiticas são aplicadas a modelos lineares.

A aproximação de modelos não lineares em modelos linearizados tem como objetivo fornecer informações qualitativas do comportamento do sistema em torno do ponto de linearização.

### O ponto de linearização

O ponto de linearização é o ponto do sistema com pequenos alterações, ou seja, os sinais envolvidos tem que possuir pequenas amplitudes. Na maioria dos casos utiliza-se o ponto de equilibrio, também conhecido como ponto de operação ou estado estacionário.
