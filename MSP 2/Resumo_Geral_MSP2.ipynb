<h1> # Linearização de modelos não lineares


- **Definição:** Linearização é o processo de encontrar uma aproximação linear de uma função num ponto específico. É utilizada no estudo de sistemas dinâmicos para avaliar a estabilidade local de um ponto de equilíbrio de um sistema de equações diferenciais não lineares.
 
- **Importância:** A maioria dos sistemas reais gera modelos não lineares, mas a maior parte das técnicas analíticas se aplica a modelos lineares. Assim, a linearização busca transformar modelos não lineares em modelos linearizados para fornecer informações qualitativas do comportamento do sistema em torno do ponto de linearização.

- **Ponto de Linearização:** Este é o ponto do sistema que sofre pequenas alterações e envolve sinais de pequenas amplitudes. Geralmente, utiliza-se o ponto de equilíbrio, também conhecido como ponto de operação ou estado estacionário.
