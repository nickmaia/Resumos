```python
# Medidas de Comprimento
conversoes_comprimento = {
    "metro": 1,
    "Bohr": 5.29177e-11,
    "ngstrm": 1e-10,
    "Mcron": 1e-6,
    "unidade": 1.002e-13,
    "polegada": 0.0254,
    "p": 0.3048,
    "jarda": 0.9144,
    "milha": 1609.344,
    "milha nutica": 1852,
    "milha geogrfica": 1855,
    "unidade astronmica": 1.496e11,
    "parsec": 3.08568e16,
    "ano-luz": 9.46073e15,
    "segundo-luz": 2.99792458e8
}

def converter_comprimento(valor, unidade_origem, unidade_destino):
    metros = valor * conversoes_comprimento[unidade_origem]
    resultado = metros / conversoes_comprimento[unidade_destino]
    return resultado

comprimento = 1
polegadas = converter_comprimento(comprimento,"metro","polegada")
print(f"{comprimento} metro(s) equivale(m) a {polegadas:.4f} polegada(s)")
```

```python
#Medidas de Massa
conversoes_massa = {
    "quilograma": 1,
    "massa do eletron": 9.10939e-31,
    "dalton": 1.66054e-27,
    "gamma": 1/1.66054e-27,
    "tonelada": 1e3,
    "libra": 0.45359237,
    "ona (avoirdupois)": 0.0283495,
    "ona (troy)": 0.0311035,
    "gro": 6.479891e-5
}

def converter_massa(valor, unidade_origem, unidade_destino):
    kilogramas = valor * conversoes_massa[unidade_origem]
    resultado = kilogramas / conversoes_massa[unidade_destino]
    return resultado


massa = 5
libras = converter_massa(massa, "quilograma", "libra")
print(f"{massa} quilograma(s) equivale(m) a {libras:.4f} libra(s)")
```

```python
#Medidas de Tempo
conversoes_tempo = {
    "segundo": 1,
    "u. a. de tempo": 2.41888e-17,
    "minuto": 60,
    "hora": 3600,
    "dia": 86400,
    "semana": 604800,
    "ms": 2592000,
    "ano": 31556952,
    "svedberg": 1e-13
}

def converter_tempo(valor, unidade_origem, unidade_destino):
    segundos = valor * conversoes_tempo[unidade_origem]
    resultado = segundos / conversoes_tempo[unidade_destino]
    return resultado


hora = 1
segundo = converter_tempo(hora, "hora", "segundo")
minuto = converter_tempo(hora, "hora", "minuto")
dia = converter_tempo(hora, "hora", "dia")

print(f"{hora} hora(s) equivale(m) a {segundo:.4f} segundo(s) ou {minuto:.4f} minuto(s) ou {dia:.4f} dia(s)")
```

```python
#Medidas de temperatura
conversoes_temperatura = {
    "Kelvin": 1,
    "grau Celsius": 1,
    "grau Fahrenheit": 5/9,
    "Rankine": 5/9
}

def converter_temperatura(valor, unidade_origem, unidade_destino):
    if unidade_origem == "grau Celsius":
        valor_kelvin = valor + 273.15
    elif unidade_origem == "grau Fahrenheit":
        valor_kelvin = (valor + 459.67) * 5/9
    else:
        valor_kelvin = valor

    if unidade_destino == "Kelvin":
        resultado = valor_kelvin
    elif unidade_destino == "grau Celsius":
        resultado = valor_kelvin - 273.15
    elif unidade_destino == "grau Fahrenheit":
        resultado = valor_kelvin * 9/5 - 459.67
    else:
        resultado = valor_kelvin * 9/5

    return resultado

temperatura = 50
kelvin = converter_temperatura(temperatura, "grau Celsius", "Kelvin")
fahrenheit = converter_temperatura(temperatura, "grau Celsius", "grau Fahrenheit")
print(f"{temperatura} grau(s) Celsius equivale(m) a {kelvin:.4f} Kelvin e {fahrenheit:.4f} grau(s) Fahrenheit")
```

### Exemplo:

Um manmetro de contrapeso com um mbolo de 1 cm de dimetro  utilizado para realizar medies precisas de presso. Em uma situao especfica, o equilbrio  alcanado com uma massa de 6,14 kg, incluindo o pisto e a plataforma. Se a acelerao da gravidade local for 9,82 m/s, qual  a presso manomtrica medida? Se a presso baromtrica  igual a 748 Torr, qual  a presso absoluta?

#### Resposta:

Vamos comear resolvendo a primeira parte do problema: encontrar a presso manomtrica medida pelo manmetro a contrapeso.

A rea do mbolo pode ser calculada usando a frmula da rea de um crculo:

A =  * (d/2)^2

Onde:
A = rea do mbolo
d = dimetro do mbolo = 1 cm = 0,01 m (convertido para metros)
 = 3,14159 (constante)

A = 3,14159 * (0,01 / 2)^2
A  7,854 * 10^(-5) m^2

Agora, podemos calcular a fora exercida pela massa no mbolo:

F = m * g

Onde:
F = fora
m = massa = 6,14 kg
g = acelerao da gravidade = 9,82 m/s

F = 6,14 kg * 9,82 m/s
F  60,25 N

A presso manomtrica (P_man) pode ser encontrada dividindo a fora pela rea:

P_man = F / A

P_man = 60,25 N / 7,854 * 10^(-5) m^2
P_man  766,720 Pa (Pascal)

Para converter a presso manomtrica em atm, usamos a converso:

1 atm = 101325 Pa

P_man(atm) = 766,720 Pa / 101325 Pa/atm
P_man(atm)  0,00757 atm

Agora, vamos encontrar a presso absoluta. Primeiro, precisamos converter a presso baromtrica dada em torr para atm:

1 atm = 760 torr

P_bar(atm) = 748 torr / 760 torr/atm
P_bar(atm)  0,98421 atm

A presso absoluta (P_abs) pode ser encontrada somando a presso manomtrica e a presso baromtrica:

P_abs = P_man + P_bar
P_abs = 0,00757 atm + 0,98421 atm
P_abs  0,99178 atm

Portanto, a presso manomtrica medida  de aproximadamente 0,00757 atm (ou 766,720 Pa), e a presso absoluta  de aproximadamente 0,99178 atm.
