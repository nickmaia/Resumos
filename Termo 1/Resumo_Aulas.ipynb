
### Processos termodinâmicos 
São mudanças de estado que ocorrem dentro do sistema. Essas mudanças podem ser isobáricas, isocóricas, isotérmicas ou adiabáticas, dependendo das variáveis termodinâmicas que permanecem constantes durante o processo.

- **Processo Isobárico:** Um processo isobárico ocorre quando a **``pressão do sistema permanece constante``** durante todo o processo. Um exemplo comum é o aquecimento de um gás dentro de um cilindro com um pistão móvel. À medida que o gás é aquecido, sua temperatura e volume aumentam, mas a pressão permanece constante devido ao ajuste do pistão.

- **Processo Isocórico (ou isovolumétrico):** Neste tipo de processo, o **``volume do sistema permanece constante``** enquanto outras variáveis termodinâmicas, como pressão e temperatura, podem mudar. Um exemplo de um processo isocórico é o aquecimento de um gás em um recipiente rígido e fechado. Como o volume do recipiente não muda, o aumento da temperatura do gás resulta em um aumento da pressão interna.

- **Processo Isotérmico:** Um processo isotérmico ocorre quando a **``temperatura do sistema permanece constante``** durante a mudança de estado. Isso geralmente implica que o sistema está em equilíbrio térmico com o ambiente. Um exemplo de um processo isotérmico é a expansão lenta e controlada de um gás em um cilindro com um pistão móvel, mantendo a temperatura constante através de um banho térmico. Nesse caso, a pressão e o volume do gás mudam, mas a temperatura não.

- **Processo Adiabático:** Um processo adiabático é aquele em que **``não há troca de calor entre o sistema e o ambiente``**. Isso pode ocorrer em sistemas bem isolados ou em processos que ocorrem muito rapidamente, não permitindo tempo suficiente para trocas de calor. Um exemplo de um processo adiabático é a compressão rápida de um gás em um cilindro com um pistão. A rápida compressão resulta em um aumento da temperatura e da pressão do gás, mas sem a troca de calor com o ambiente.

---


### Classificação do Estado de um Sistema:

- **Estado Estacionário ou Permanente:** Os valores de todas as variáveis do processo (temperatura, pressão, volume, vazões, etc.) ``não variam com o tempo``.
- **Estado Transiente ou Não-Estacionário:** Pelo menos uma variável do processo ``varia com o tempo``.

**Equilíbrio Termodinâmico:** É o estado em que as variáveis termodinâmicas possuem valores uniformes e estacionários (``não variam com o tempo``) em todo o sistema. É uma condição na qual não há mudanças espontâneas nas propriedades do sistema e, portanto, a ``energia e a matéria`` estão distribuídas de ``forma estável``.

---

### Fases 

Na termodinâmica, a "fase" de uma substância refere-se ao estado físico em que ela se encontra, como sólida, líquida ou gasosa. 

As fases mais comuns são:

- **Sólida**: Nesta fase, as partículas (átomos, íons ou moléculas) estão organizadas em uma estrutura cristalina ou amorfa, com forte interação entre as partículas. Os sólidos têm forma e volume definidos e geralmente possuem alta densidade e baixa compressibilidade.
- **Líquida**: Na fase líquida, as partículas estão mais próximas umas das outras do que na fase gasosa, mas menos organizadas do que na fase sólida. Os líquidos têm volume definido, mas sua forma depende do recipiente em que estão contidos. A densidade dos líquidos é geralmente menor que a dos sólidos e maior que a dos gases, e possuem maior compressibilidade que os sólidos, mas menor que os gases.
- **Gasosa**: Na fase gasosa, as partículas estão dispersas e se movem livremente, com interações fracas entre elas. Os gases não têm forma nem volume definidos, ocupando todo o espaço disponível no recipiente que os contém. A densidade dos gases é menor do que a dos líquidos e sólidos, e possuem alta compressibilidade.

---

### Mudança de fases
As mudanças de fase ocorrem quando uma substância passa de um estado físico para outro devido a alterações em variáveis termodinâmicas, como pressão e temperatura.



#### Exemplo a Água:
A **fusão** (passagem do estado sólido para o líquido) ocorre quando a temperatura do sólido atinge seu ponto de fusão sob uma pressão constante. Outras mudanças de fase incluem a **vaporização** (líquido para gasoso), **condensação** (gasoso para líquido), **sublimação** (sólido para gasoso) e **deposição** (gasoso para sólido).

---

### Equilíbrio Térmico



Quando os primeiros métodos de quantificação da temperatura foram desenvolvidos, tornou-se possível estabelecer o que hoje chamamos de **equilíbrio térmico** em um sistema. Observou-se que, quando ``corpos quentes eram colocados em contato com corpos frios, o corpo mais quente se resfriava enquanto o corpo mais frio se aquecia``. Eventualmente, ``os dois corpos alcançavam uma temperatura comum, onde não se percebia mais nenhuma mudança na sensação térmica``. Esse estado é conhecido como equilíbrio térmico.

## # Estado Termodinâmico e Funções de Estado
O conjunto de valores das propriedades termodinâmicas de um sistema define o seu estado termodinâmico. As funções de estado são variáveis que dependem apenas das condições atuais do sistema, não de sua história passada.

Por exemplo, a ``energia interna é uma função de estado porque depende apenas das condições atuais do sistema``, e não de como o sistema chegou àquele estado. Por outro lado, ``o calor (Q) e o trabalho (W) não são funções de estado, pois dependem da trajetória que o sistema seguiu para atingir o estado atual``.

Para um sistema fechado, diferentes processos podem levar ao mesmo estado termodinâmico, mas as quantidades de calor e trabalho envolvidas em cada processo serão diferentes. No entanto, a soma de Q e W será sempre a mesma, independentemente do processo. Isso é expresso pela primeira lei da termodinâmica:

### $$\Delta U = Q + W$$

## # Equilíbrio Termodinâmico
Existem três tipos de equilíbrio termodinâmico: térmico, mecânico e químico. Um sistema está em equilíbrio térmico quando todas as partes têm a mesma temperatura. Está em equilíbrio mecânico quando não há mudanças macroscópicas no movimento e em equilíbrio químico quando a composição química não está mudando.

``Um sistema está em equilíbrio termodinâmico quando está simultaneamente em equilíbrio térmico, mecânico e químico. Este é o estado mais estável que um sistema pode atingir.``

## # A Regra das Fases
A Regra das Fases fornece o número mínimo de variáveis intensivas que devem ser especificadas para determinar completamente o estado de um sistema. Por exemplo, para um sistema de uma única substância, a regra das fases afirma que precisamos especificar duas variáveis intensivas (como pressão e temperatura) para determinar completamente o estado do sistema.


### $$F = 2 - \pi + N$$

Onde:
- **F** é o número de graus de liberdade.
- **N** é o número de componentes.
- **$\pi$** é o número de fases.

## Processos Termodinâmicos: Reversíveis, Irreversíveis e Mecanicamente Reversíveis

### Processos Reversíveis

Os processos termodinâmicos reversíveis são aqueles que podem ser revertidos por pequenas alterações no ambiente. Em tais processos:

- O sistema nunca se desvia significativamente do equilíbrio;
- Ocorre por meio de uma série de estágios de equilíbrio;
- Pode ser revertido em qualquer ponto, restaurando o estado inicial tanto do sistema quanto do ambiente.

No entanto, na prática, processos perfeitamente reversíveis são raros, pois exigem mudanças infinitesimais ocorrendo em um período de tempo infinitamente longo.

### Processos Irreversíveis

Os processos termodinâmicos irreversíveis são aqueles que não podem ser revertidos por pequenas alterações no ambiente. Possuem uma direção definida, a "seta do tempo", que indica para onde o processo segue e de onde não pode voltar. Exemplos comuns de processos irreversíveis incluem o envelhecimento dos seres vivos, um copo de vidro que cai e se quebra, e um pão que é assado.

Importante ressaltar que um sistema passando por um processo irreversível não está impedido de retornar ao seu estado inicial. Porém, se o sistema retornar ao estado inicial, o mesmo não será possível para a sua vizinhança. Causas que tornam um processo irreversível incluem atrito, expansão não resistida, mistura de duas substâncias diferentes, troca de calor com diferença de temperatura, entre outros.

### Processos Mecanicamente Reversíveis

Todos os processos reais são acompanhados, em maior ou menor escala, por algum efeito dissipativo. Os processos que satisfazem duas condições - equilíbrio interno do sistema (caracterizado pela uniformidade das propriedades internas do sistema) e o fato do sistema nunca se afastar mais que infinitesimalmente do equilíbrio mecânico com a sua vizinhança - são denominados mecanicamente reversíveis.

## # Processos Termodinâmicos a Volume e Pressão Constante

### Processos a Volume Constante

Em um processo a volume constante (também chamado de isocórico), o volume do sistema não muda. Se considerarmos um sistema fechado contendo n moles de um fluido, o balanço de energia para esse sistema passando por um processo mecanicamente reversível é descrito pela primeira lei da termodinâmica:

### $$d(nU)=dQ+dw$$

Onde:

- $d(nU)$ é a variação diferencial na energia interna do sistema,
- $dQ$ é o calor diferencial transferido para o sistema,
- $dw$ é o trabalho diferencial realizado pelo sistema.

Se o processo for mecanicamente reversível, o trabalho diferencial é dado por:

### $$dw = -Pd(nV)$$

Substituindo essa expressão na primeira lei da termodinâmica, obtemos:

### $$d(nU)=dQ-Pd(nV)$$

Para um processo a volume constante, $d(nV) = 0$ (já que o volume não muda), portanto:

### $$dQ=d(nU)$$

Integrando, obtemos:

### $$Q=n \Delta U$$

Assim, em um sistema fechado passando por um processo mecanicamente reversível a volume constante, o calor trocado é igual à variação na energia interna.

### Processos a Pressão Constante

Para um processo a pressão constante (ou isobárico), explicitando $dQ$ na equação diferencial do balanço de energia, obtemos:

### $$dQ=d(nU) + Pd(nV)$$

Se a pressão $P$ é constante, isso simplifica para:

### $$dQ = d(nU) + d(nPV) = d[n(U+PV)]$$

Definimos entalpia $H$ como $H = U + PV$, então:

### $$dQ=d(nH)$$

Integrando, obtemos:

### $$Q= n \Delta H$$

Assim, em um sistema fechado passando por um processo mecanicamente reversível a pressão constante, o calor trocado é igual à variação na entalpia.

### A Entalpia

A variação da entalpia é dada por:

### $$dH = dU + d(PV)$$

Ou de forma integrada:

### $$\Delta H = \Delta U + \Delta (PV)$$

Essas relações fornecem as bases para análises de processos a volume e pressão constantes em termodinâmica.

## # Capacidade Calorífica

A capacidade calorífica (C)  ~é uma propriedade extensiva, pois ela depende da quantidade de substância presente.~  É a medida da quantidade de calor necessário para aumentar a temperatura de um determinado objeto ou substância por uma determinada quantidade. Ela é comumente expressa em termos de quantidade de calor por grau de temperatura (como em joules por Kelvin ou calorias por grau Celsius). 

Ela pode ser representada de diversas formas:

- **Capacidade calorífica específica (Ce)**: É a capacidade calorífica por unidade de massa da substância. Ela é expressa em joules por Kelvin por grama ($\frac{J}{(K·g)}$).

    $$ ce = C/m $$

- **Capacidade calorífica molar (Cm)**: É a capacidade calorífica por mol da substância. Ela é expressa em joules por Kelvin por mol ($\frac{J}{(K·mol)}$).

    $$ cm = C/n $$

### Capacidade Calorífica a Volume Constante (Cv)

A capacidade calorífica a volume constante (Cv) representa a quantidade de calor necessário para aumentar a temperatura de uma substância em um grau Kelvin ou Celsius a volume constante. A capacidade calorífica a volume constante é uma função de estado, pois seu valor depende apenas do estado inicial e final e não do caminho percorrido para chegar ao estado final.

A capacidade calorífica a volume constante é dada por:

$$ Cv = \left( \frac{{dU}}{{dT}} \right)_V $$

Ou seja, a variação da energia interna (U) pode ser calculada por:

$$ dU = Cv \, dT $$

$$ \Delta U = \int_{T1}^{T2} Cv \, dT $$

Em um processo mecanicamente reversível a volume constante, a quantidade de calor trocado é igual à variação de energia interna:

$$ Q = n \Delta U = n \int_{T1}^{T2} Cv \, dT $$

### Capacidade Calorífica a Pressão Constante (Cp)

A capacidade calorífica a pressão constante (Cp) representa a quantidade de calor necessário para aumentar a temperatura de uma substância em um grau Kelvin ou Celsius a pressão constante. 

A capacidade calorífica a pressão constante é dada por:

$$ Cp = \left( \frac{{dH}}{{dT}} \right)_P $$

Ou seja, a variação de entalpia (H) pode ser calculada por:

$$ dH = Cp \, dT $$

$$ \Delta H = \int_{T1}^{T2} Cp \, dT $$

Em um processo mecanicamente reversível a pressão constante, a quantidade de calor trocado é igual à variação de entalpia:

$$ Q = n \Delta H = n \int_{T1}^{T2} Cp \, dT $$

## # Balanços de Massa e Energia em Sistemas Abertos

A formulação de um balanço de massa para um volume de controle (VC) em um sistema aberto é expressa pela equação de continuidade. A equação de continuidade é uma expressão matemática da conservação de massa para um volume de controle. 

De acordo com a lei de conservação de massa:


#### $$\frac{d (m)_{vc}}{dt} + \Delta (\dot{m})_{corr} = 0$$


onde:

- $m_{vc}$ é a massa dentro do volume de controle,
- $\dot{m}$ é a vazão mássica (massa por unidade de tempo),



O termo $\Delta (\dot{m})_{corr}$ é definido como a diferença entre as vazões mássicas que entram e saem do volume de controle, ou seja:


#### $$\Delta (\dot{m})_{corr} = \dot{m}_3 - \dot{m}_1 - \dot{m}_2$$


onde:
- $\dot{m}_1$, $\dot{m}_2$ são as vazões mássicas entrando no volume de controle, e
- $\dot{m}_3$ é a vazão mássica saindo do volume de controle.


Quanto às medidas de escoamento, elas são quantidades importantes que caracterizam o movimento de fluidos. 

A vazão mássica ($\dot{m}$) é a quantidade de massa que flui através de uma seção por unidade de tempo. Ela pode ser calculada pela fórmula:


#### $$\dot{m} = ρAu$$

A vazão molar ($\dot{n}$) é a quantidade de substância que flui através de uma seção por unidade de tempo. Ela pode ser calculada pela fórmula:


#### $$\dot{n} = \frac{\dot{m}}{M}$$


Agora substituindo a equação (2) em (1), obtemos a equação da continuidade:

#### $$\frac{d (ρV)_{vc}}{dt} + \Delta (ρuA)_{corr} = 0$$

onde:
- $V_{vc}$ é o volume do volume de controle.
- $\rho$ é a densidade do fluido,
- $u$ é a velocidade média do fluido, e
- $A$ é a área da seção transversal do escoamento.


Em estado estacionário, a massa dentro do volume de controle não varia com o tempo. Portanto, a equação da continuidade se reduz a:

#### $$\Delta (\rho uA)_{corr} = 0$$

Nesta condição, a soma das vazões mássicas que entram no volume de controle é igual à soma das vazões mássicas que saem do volume de controle. Ou seja, não há acúmulo de massa dentro do volume de controle.

# Balanço de Energia Geral

A equação geral para o balanço de energia em um volume de controle é dada por:

#### $$[\frac{d(mU)_{vc}}{dt} = -\Delta[(U+u^2/2+gz)\dot{m}] + \dot{Q} + \dot{W}_{e}]$$  (1)

Onde:
- $(U)$ é a energia interna do sistema,
- $(u)$ é a velocidade do fluido,
- $(g)$ é a aceleração da gravidade,
- $(z)$ é a altura em relação a um ponto de referência,
- $(\dot{m})$ é a taxa de variação de massa,
- $(\dot{Q})$ é a taxa de transferência de calor e
- $(\dot{W}_{e})$ é a taxa de trabalho realizado pelo sistema.

A taxa de trabalho é dada por:

#### $$[\dot{W}_{e} = \dot{W} - \Delta[(PV)\dot{m}]_{corr}]$$ (2)

Substituindo a equação (2) na equação (1) temos:

#### $$[\frac{d(mU)_{vc}}{dt} = -\Delta[(U+u^2/2+gz)\dot{m}]_{corr} + \dot{Q} + \dot{W} - \Delta[(PV)\dot{m}]_{corr}]$$

Considerando que a entalpia \(H\) é dada por \(U+PV\), podemos escrever a equação como:

#### $$[\frac{d(mU)_{vc}}{dt} + \Delta[(H+u^2/2+gz)\dot{m}]_{corr} = \dot{Q} + \dot{W}]$$

## Simplificações

Em muitos casos práticos, a variação da energia cinética e potencial do fluido nas correntes pode ser negligenciada. Portanto, a equação simplificada torna-se:

#### $$[\frac{d(mU)_{vc}}{dt} + \Delta[H\dot{m}]_{corr} = \dot{Q} + \dot{W}]$$

## Caso de Escoamento Permanente em Estado Estacionário

No caso de um escoamento permanente em estado estacionário, as parcelas variantes com o tempo são nulas, significando que as propriedades do volume de controle não variam com o tempo. 

A única forma de trabalho que pode ocorrer é o trabalho de eixo. Portanto, a aplicação desta equação ocorre em processos com escoamento estacionário em estado estacionário.

Neste caso, a equação se torna:

#### $$[\Delta[(H+u^2/2+gz)\dot{m}]_{corr} = \dot{Q} + \dot{W}_{e}]$$

Para um processo com uma única entrada e saída, a equação fica:

#### $$[\Delta(H+u^2/2+gz)\dot{m} = \dot{Q} + \dot{W}_{e}]$$

Dividindo pela vazão mássica temos:

#### $$[\Delta H + \Delta u^2/2 + g\Delta z = W_{e} + Q]$$

Caso a energia cinética e a energia potencial possam ser negligenciadas, a equação simplifica para:

#### $$[\Delta H = W_{e} + Q]$$

Obs: Muitas vezes $(W_{e})$ é simplificado para $(W)$.

# Escoamento Permanente

Escoamento permanente, ou estacionário, é aquele em que a velocidade e a pressão em um determinado ponto não variam com o tempo. A velocidade e a pressão podem variar de um ponto para outro do fluxo, mas se mantêm constantes em cada ponto do espaço, em qualquer instante de tempo. Assim, a pressão e a velocidade em um ponto são funções das coordenadas do ponto e não dependentes do tempo. No escoamento permanente, o fluxo do fluido é dito "estável".
