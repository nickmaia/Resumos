## Reações de Substituição Nucleofílica: SN1 e SN2

As reações de substituição nucleofílica são fundamentais na química orgânica, onde um nucleófilo substitui um grupo de saída em uma molécula. Existem dois mecanismos predominantes para essas reações: SN1 e SN2.

## # SN1 (Substituição Nucleofílica Unimolecular)

O mecanismo SN1 é caracterizado pela formação de um carbocátion, seguido pelo ataque de um nucleófilo e, finalmente, pela ação de uma base. A reação é unimolecular, o que significa que a velocidade da reação depende apenas da concentração do haleto de alquila.

##### Mecanismo
1. **Formação do Carbocátion**: A molécula se dissocia, liberando o grupo de saída e formando um carbocátion intermediário.
2. **Ataque Nucleofílico**: O nucleófilo ataca o carbocátion, gerando o produto final.

##### Características
- **Cinética**: Primeira ordem.
- **Estereoquímica**: Pode resultar em uma mistura racêmica.
- **Fatores Influenciadores**: Estabilidade do carbocátion e capacidade do grupo de saída.

$$\text{R-X} \rightarrow \text{R}^{+} + \text{X}^{-} \rightarrow \text{R-Nu}$$


#### Exemplo

###### **Explicação:** Inicialmente, o halogênio sai sozinho, formando um carbocátion. Posteriormente, um nucleófilo ataca o carbocátion, resultando em uma molécula protonada. Finalmente, uma base destrona o hidrogênio, levando ao produto final.
###### - **Carbocátion:** É uma molécula que possui um carbono com carga positiva. No mecanismo SN1, o carbocátion é formado quando o halogênio sai da molécula.
###### - **Nucleófilo:** No contexto da reação SN1, o nucleófilo geralmente é uma molécula neutra que ataca o carbocátion. O nucleófilo compartilha seu par de elétrons com o carbocátion, resultando em uma molécula protonada.
###### - **Base:** A base é responsável por destronar o hidrogênio da molécula protonada, levando à formação do produto final.
    



#### Velocidade da Reação [SN1]
Em reações SN1, a velocidade da reação é determinada principalmente pelo grupo de saída e pela estabilidade do carbocátion formado. 



######  **Obs:** Em reações SN1, o nucleófilo geralmente é o solvente (*Solvólise*). A reatividade do nucleófilo não afeta a velocidade da reação, pois ele entra em ação após a etapa determinante da velocidade.

#### Estereoquímica de reações SN1

Na reação SN1, é possível obter produtos com a mesma configuração do reagente ou com configuração invertida. Isso ocorre porque o nucleófilo pode atacar qualquer lado do carbocátion planar. Se o nucleófilo atacar pelo mesmo lado de onde o íon saiu, o produto terá a mesma configuração que o reagente. Se atacar pelo lado oposto, o produto terá a configuração invertida.

- **Determinação da Configuração:** A configuração de uma molécula pode ser determinada usando regras de prioridade baseadas no número atômico dos átomos ligados ao carbono central. O bromo, por exemplo, tem uma prioridade maior do que o carbono devido ao seu maior número atômico.



######  **Direção:** Observe a direção dos três substituintes restantes, do de maior prioridade para o de menor:
######  - Se a direção for ***horária***, a configuração é R.
######  - Se a direção for ***anti-horária***, a configuração é S.


## # SN2 (Substituição Nucleofílica Bimolecular)
O mecanismo SN2 envolve uma colisão produtiva em que o nucleófilo ataca o carbono pelo lado oposto ao grupo de saída. O grupo de saída bloqueia a aproximação do nucleófilo pela frente da molécula.

##### Mecanismo
1. **Ataque Concertado**: O nucleófilo ataca o carbono simultaneamente à saída do grupo de saída, resultando em uma inversão de configuração.

##### Características
- **Cinética**: Segunda ordem.
- **Estereoquímica**: Sempre ocorre inversão de configuração.
- **Fatores Influenciadores**: Impedimento estérico, força do nucleófilo e capacidade do grupo de saída.

$$\text{Nu}^{-} + \text{R-X} \rightarrow \text{R-Nu} + \text{X}^{-}$$

Onde R é um grupo alquila e X é um halogênio (Br, Cl, I).

######  **Obs:** O fluor (F) não é considerado, pois tem uma velocidade muito baixa em relação ao outros halogênios.

#### Exemplos

1. Nucleófilo (Nu) ataca um carbono ligado a um bromo (Br). O bromo é liberado como grupo de saída, e o nucleófilo substitui o bromo.


   
2. Nucleófilo (Nu) ataca um carbono ligado a um cloro (Cl). O cloro é liberado, e o nucleófilo toma seu lugar.




#### Velocidade Relativa da reação [SN2]




######  **Obs:** Quanto mais fraca for a base maior será a velocidade da reação.

#### Inversão de configuração de uma reação SN2

Durante uma reação SN2, o carbono que sofre a substituição inverte sua configuração. Esta inversão é comparada à forma como um guarda-chuva se inverte ao ser exposto a uma ventania forte.

Exemplo: A reação do 2-bromobutano (configuração R) com hidróxido resultará em um produto com configuração S.



######  **Obs:** Se um reagente tem configuração R, o produto da reação SN2 terá configuração S, e vice-versa.

######  **Carbono Assimétrico:** É um carbono ligado a quatro grupos diferentes. A presença de um carbono assimétrico em uma molécula a torna quiral, o que significa que ela pode existir em duas formas espaciais diferentes (isômeros ópticos).

#### Comparação entre SN1 e SN2

| Propriedade | SN1 | SN2 |
|-------------|-----|-----|
| Ordem da reação | 1ª ordem | 2ª ordem |
| Intermediários | Carbocátion | Nenhum |
| Estereoquímica | Pode formar mistura racêmica | Inversão de configuração |
| Ordem de Atividade | Terciário > Secundário > Primário | Metil > Primário > Secundário |
| Determinação do Mecanismo Predominante | Necessita de um solvente polar prótico e um nucleófilo pobre (neutro, como água ou álcool) | Necessita de um solvente polar aprótico e um nucleófilo rico (com carga negativa) |

### Diferença entre Base e Nucleófilo

Embora muitas vezes sejam usados como sinônimos, base e nucleófilo têm diferenças. A base está mais relacionada à capacidade de compartilhar elétrons, enquanto o nucleófilo está relacionado à capacidade de atacar um átomo deficiente em elétrons.

**Exemplo:** Uma base forte, rica em elétrons, geralmente é um bom nucleófilo porque tem a capacidade de atacar átomos deficientes em elétrons. No entanto, existem exceções, e nem toda base forte é necessariamente um bom nucleófilo.



---

### Estabilidade de Carbocátions

A estabilidade de um carbocátion está diretamente relacionada ao número de substituintes ligados ao carbono carregado positivamente. Quanto mais substituintes, maior a estabilidade.

- **Carbocátions Terciários:** São os mais estáveis porque possuem três substituintes ligados ao carbono carregado positivamente. Esses substituintes doam densidade eletrônica, estabilizando o carbocátion.

- **Carbocátions Secundários:** São menos estáveis que os terciários, mas mais estáveis que os primários. Possuem dois substituintes ligados ao carbono carregado positivamente.

- **Carbocátions Primários:** São os menos estáveis entre os três, possuindo apenas um substituinte ligado ao carbono carregado positivamente.




### Rearranjo de carbocátion

Se um carbocátion terciário for mais estável do que um secundário ou primário, pode ocorrer um rearranjo. O rearranjo envolve a migração de um hidrogênio ou de um grupo metila para transformar um carbocátion menos estável em um mais estável.

**Rearranjo 1,2 Hidreto:** Um hidrogênio migra de um carbono adjacente para o carbono carregado positivamente, resultando na formação de um carbocátion mais estável.



**Rearranjo 1,2 de Metila:** Um grupo metila migra de um carbono adjacente para o carbono carregado positivamente, também resultando na formação de um carbocátion mais estável.
