## # Mecanismos de Transferência de Calor

A tranferência de calor é a energia térmica em transito (q) devido a uma diferença de temperaturas no espaço.

Esta tranferência de calor (q) se dá por:

### 1. Condução

A condução por sua vez é a transferência de energia **(q)** por meio de um sólido ou fluido estacionário:



A condução térmica, como já definido, pode ser quantificada pela ``Lei de Fourier``. Assim foi determinado a taxa e o fluxo para a transferência de calor na direção X:

### Taxa:
### $$q_{x}= -K.A \frac{dT}{dx}$$

### Fluxo:
### $$\dot{q_{x}}= -K \frac{dT}{dx}$$

Unidades:

- $\frac{dT}{dx}$: Gradiente de temperatura [$\frac{K}{m}$]
- K: Condutividade térmica do meio [$\frac{W}{m.K}$ ou $\frac{W}{m.°C}$]
- A: área [m²]
- $q_{x}$: [W] ou [J/s]
- $\dot{q_{x}}$: [W/m²] ou [J/s.m²]

Determinando $\bar{K}$ médio utilizando o fluxo da transferência de calor:

### $$\int \dot{q}_{x} dx =  - \int_{T1}^{T2} K . dT = \bar{K}$$

Tal que, se o erro for menor que **10%** o $\bar{K}$ médio é válido.

### $$\bar{K}=\frac{K1+K2}{2}$$
